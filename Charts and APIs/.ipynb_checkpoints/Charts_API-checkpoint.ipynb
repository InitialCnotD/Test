```python
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt 
response_API = requests.get('https://api.covid19api.com/live/country/singapore')
print(response_API.status_code)
```

```python
data = response_API.text
df = pd.read_json(data)
df
```

```python
plt.plot(df['Date'], df['Confirmed'], color='k', label='Confirmed')
plt.title('Confirmed Case in Singapore')
plt.legend()
plt.show()
```

```python
plt.plot(df['Date'], df['Deaths'], color='r', label='Deaths')
plt.title('Death Case in Singapore')
plt.legend()
plt.show()
```

```python
plt.plot(df['Date'], df['Recovered'], color='b', label='Recovered')
plt.plot(df['Date'], df['Active'], color='g', label='Active')
plt.title('Recovered and Active Case in Singapore')
plt.legend()
plt.show()
```
